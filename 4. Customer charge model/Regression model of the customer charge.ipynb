    - Demographic info about customers – gender, age range, and if they have partners and dependents
- The used dataset was obtained in https://www.kaggle.com/blastchar/telco-customer-churn

#### 1.3.1 Machine learning models auto-training and evaluate

```python
def models_score(models, X_train, y_train, X_test, y_test):
  results = []
  names = []
  for name, model in models:
    names.append(name)
    # model training
    time1 = datetime.datetime.now()
    model.fit(X_train, y_train)
    time2 = datetime.datetime.now()
    # training time
    duration=time2-time1 
    # df
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    results.append({ 'Modelo': name,
                    'R2-score (train)': model.score(X_train, y_train),
                    'Explained variance score (train)': explained_variance_score(y_train, y_pred_train),
                    'Max error (train)': max_error(y_train, y_pred_train),
                    'MAE (train)': mean_absolute_error(y_train, y_pred_train),
                    'MSE (train)': mean_squared_error(y_train, y_pred_train),
                    'MSLE (train)': mean_squared_log_error(y_train, y_pred_train),
                    'MedAE (train)': median_absolute_error(y_train, y_pred_train),
                    'R2-score (test)': model.score(X_test, y_test),
                    'Explained variance score (test)': explained_variance_score(y_test, y_pred_test),
                    'Max error (test)': max_error(y_test, y_pred_test),
                    'MAE (test)': mean_absolute_error(y_test, y_pred_test),
                    'MSE (test)': mean_squared_error(y_test, y_pred_test),
                    'MSLE (test)': mean_squared_log_error(y_test, y_pred_test),
                    'MedAE (test)': median_absolute_error(y_test, y_pred_test),
                    'Duration (s)':  duration.total_seconds() 
              })
  display(pd.DataFrame(results).sort_values(by='R2-score (test)', ascending = False))
```

### 2.2 Train and test split

```python
train_columns = list(data.columns)
train_columns = [e for e in train_columns if e not in ('Churn', 'tenure_group_1', 'tenure_group_2', 'tenure_group_3', 'tenure_group_4', 'tenure_group_5+', 'TotalCharges', 'MonthlyCharges')]
```

```python
# original train/test dataset
X_train, X_test, y_train, y_test = train_test_split(
    data[train_columns], data['MonthlyCharges'], test_size=0.3, random_state=seed, stratify=data['Churn']
                                                    )

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

### 2.3 Normalization

```python
X_train_bckp = X_train.copy()
X_test_bckp = X_test.copy()
```

```python
mm_scaler = preprocessing.MinMaxScaler() # scaler
# original dataset
X_train = mm_scaler.fit_transform(X_train)
X_test = mm_scaler.transform(X_test)
```

## 3. Baseline model

A baseline is a method that uses heuristics, simple summary statistics, randomness, or machine learning to create predictions for a dataset. These predictions can be used to measure the baseline's performance (e.g., r2-score). 

A model with greater r2-score than the one found for the baseline model will be a better model than if we follow simple rules for prediction.

In this case, It will be used a dummy regressor that uses the mean value strategy. 

```python
models = []
models.append(('Dummy', DummyRegressor(strategy="mean")))
models_score(models, X_train, y_train, X_test, y_test)
```

So, a model with an r2-score higher than -0.001893 is better than a simple mean algorithm.

#### 4.9.1 Most important features

```python
print('Accordingly the LR model, the most important features are: \n')
d = {'Feature':train_columns,'Importance':lr.coef_.tolist()} 
df_explic = pd.DataFrame(d, columns=['Feature','Importance'])
df_explic['signal'] = df_explic['Importance'].apply(lambda x: '+' if x > 0 else '-')
df_explic['Importance'] = np.abs(df_explic['Importance'])
df_explic.sort_values(by='Importance', ascending=False)
```

This table presents all the features and the importance attributed to them by the linear regression. 
In addition, the sign column shows the direction of this importance, that is, a positive sign is related to directly proportional attributes in relation of the monthly charge (the higher, the greater the charge), while a minus sign is related to inversely proportional attributes (the smaller, the greater the charge).

The same information obtained in this table can be displayed in a plot for better visualization.

```python
plt.figure(figsize=(20,8))
sns.barplot(x="Importance", y="Feature", data=df_explic.sort_values(by='Importance', ascending=False)[:15])
plt.title('Top 15 features', fontsize=16)
plt.xlabel('Importance', fontsize=13)
plt.ylabel('Feature', fontsize=13)
plt.show()
```

In this experiment, a regression model was created to predict the monthly charge paid by the customer. This model is useful to be used with the churn propensity model, in the final table of propensity, where it can be compared if the value of the customer prone to churn is outdated (greater) than that of customers who are not prone.

In this way, we were also able to understand the impact on cost that each service purchased has, as well as their combination in bundle.
